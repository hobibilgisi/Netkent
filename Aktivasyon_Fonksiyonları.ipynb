```python
import math                       # math ktphanesi import edildi
%matplotlib inline
                                  # oluturulacak grafiin dahili ekranda gsterilmesini salar
import matplotlib.pyplot as plt   # matplotlip ktphanesinin pyplot modl import edildi ve plt olarak adlandrld
import numpy as np                # numpy ktphanesi import edildi ve np olarak adlandrld
import matplotlib                 # matplotlip ktphanesi import edildi
import matplotlib.backends        # matplotlip ktphanesinin backends modl import edildi
#------------------------------------------
#aktivasyon fonksiyonlar, activation functions
# aktivasyon fonksiyonlarnn python kodlar birer fonksiyon olarak tanmland ve k parametreleri belirlendi

def step(x):
    return np.array(x > 0, dtype=np.int64)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, katsayi=0.06):
  return np.maximum(katsayi*x,x)

def tanh(x):
    return np.tanh(x)

def swish(x):
    return x*(1 / (1 + np.exp(-x)))

x = np.arange(-2., 3., 0.001)               # grafiin boyutlar ve hassasiyet biktar belirlendi
#------------------------------------------
# fonksiyonlar arld ve kendi adlarnda belirlenen deikenlere atand

step=step(x)
sigmoid=sigmoid(x)
relu=relu(x)
leaky_relu=leaky_relu(x-0.04)
tanh=tanh(x)
swish=swish(x)

#------------------------------------------
#grafik izimleri, plotting

plt.figure()                # mtplotlib ktphanesinde grafik figr oluturulur
plt.xlabel("girdiler")      # grafiin x ve y eksenlerine isim verildi
plt.ylabel("ktlar")
plt.grid(True)              # grafiin zgaras aktifletirildi
# grafikte gsterilecek ktlarn ad, rengi ve izgi kalnlklar belirlendi
plt.plot(x,step, label="Step", color='C0', lw=1)
plt.plot(x,sigmoid, label="Sigmoid", color='C1', lw=1)
plt.plot(x,relu, label="ReLU", color='C2', lw=1)
plt.plot(x, leaky_relu, label="Leaky ReLU", color='C5', lw=2)
plt.plot(x,tanh, label="TanH", color='C3', lw=1)
plt.plot(x,swish, label="swish", color='C4', lw=1)
plt.legend()                # grafiin lagent aktifletirildi
```
