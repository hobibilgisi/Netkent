```python
import math                       # math kütüphanesi import edildi
%matplotlib inline
                                  # oluşturulacak grafiğin dahili ekranda gösterilmesini sağlar
import matplotlib.pyplot as plt   # matplotlip kütüphanesinin pyplot modülü import edildi ve plt olarak adlandırıldı
import numpy as np                # numpy kütüphanesi import edildi ve np olarak adlandırıldı
import matplotlib                 # matplotlip kütüphanesi import edildi
import matplotlib.backends        # matplotlip kütüphanesinin backends modülü import edildi
#------------------------------------------
#aktivasyon fonksiyonları, activation functions
# aktivasyon fonksiyonlarının python kodları birer fonksiyon olarak tanımlandı ve çıkış parametreleri belirlendi

def step(x):
    return np.array(x > 0, dtype=np.int64)

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def leaky_relu(x, katsayi=0.06):
  return np.maximum(katsayi*x,x)

def tanh(x):
    return np.tanh(x)

def swish(x):
    return x*(1 / (1 + np.exp(-x)))

x = np.arange(-2., 3., 0.001)               # grafiğin boyutları ve hassasiyet biktarı belirlendi
#------------------------------------------
# fonksiyonlar çağrıldı ve kendi adlarında belirlenen değişkenlere atandı

step=step(x)
sigmoid=sigmoid(x)
relu=relu(x)
leaky_relu=leaky_relu(x-0.04)
tanh=tanh(x)
swish=swish(x)

#------------------------------------------
#grafik çizimleri, plotting

plt.figure()                # mtplotlib kütüphanesinde grafik figürü oluşturulur
plt.xlabel("girdiler")      # grafiğin x ve y eksenlerine isim verildi
plt.ylabel("çıktılar")
plt.grid(True)              # grafiğin ızgarası aktifleştirildi
# grafikte gösterilecek çıktıların adı, rengi ve çizgi kalınlıkları belirlendi
plt.plot(x,step, label="Step", color='C0', lw=1)
plt.plot(x,sigmoid, label="Sigmoid", color='C1', lw=1)
plt.plot(x,relu, label="ReLU", color='C2', lw=1)
plt.plot(x, leaky_relu, label="Leaky ReLU", color='C5', lw=2)
plt.plot(x,tanh, label="TanH", color='C3', lw=1)
plt.plot(x,swish, label="swish", color='C4', lw=1)
plt.legend()                # grafiğin lagentı aktifleştirildi
```
